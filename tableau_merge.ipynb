```python
import pandas as pd
import glob
import datetime as dt
```

```python
path = r'2017' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame_2017 = pd.concat(li, axis=0, ignore_index=True, sort=False)
frame_2017
```

```python
path = r'2019' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame_2019 = pd.concat(li, axis=0, ignore_index=True, sort=False)
frame_2019
```

```python
merge1 = frame_2017.append(frame_2018, sort=False)
merge1
```

```python
merge2 = merge1.append(frame_2019, sort=False)
merge2
```
